### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
# Função para formatar no padrão brasileiro
def formato_brasileiro(valor):
    valor_formatado = f"{valor:,.2f}"
    valor_formatado = valor_formatado.replace(",", "X").replace(".", ",").replace("X", ".")
    return valor_formatado

# Calculando o faturamento total da Loja 1
faturamento_total_loja1 = loja["Preço"].sum()

# Exibindo resultado formatado
print(f"Faturamento Total da Loja 1: R$ {formato_brasileiro(faturamento_total_loja1)}")
```

```python
loja2.head()
```

```python

# Calculando o faturamento total da loja 2
faturamento_total_loja2 = loja2["Preço"].sum()

print(f"Faturamento Total da Loja 2: R$ {formato_brasileiro(faturamento_total_loja2)}")
```

```python
loja3.head()
```

```python
# Calculando o faturamento total da loja 3
faturamento_total_loja3 = loja3["Preço"].sum()
print(f"Faturamento Total da Loja 3: R$ {formato_brasileiro(faturamento_total_loja3)}")
```

```python
loja4.head()
```

```python
# Calculando o faturamento total da loja 4
faturamento_total_loja4 = loja4["Preço"].sum()
print(f"Faturamento Total da Loja 4: R$ {formato_brasileiro(faturamento_total_loja4)}")
```

# 2. Vendas por Categoria

```python
#Loja 01:
# Agrupando por categoria e contando quantidade de vendas
vendas_por_categoria_produto = loja.groupby("Categoria do Produto")["Preço"].count().reset_index()

# Renomeando a coluna para ficar mais claro
vendas_por_categoria_produto = vendas_por_categoria_produto.rename(columns={"Preço": "Quantidade_vendida"})

# Exibindo o resultado
print (f"Resultado da loja 01:\n \n {vendas_por_categoria_produto}")
```

```python
#Loja 02
# Agrupando por categoria e contando quantidade de vendas
vendas_categoria_produto_LJ02 = loja2.groupby("Categoria do Produto")["Preço"].count().reset_index()

# Renomeando a coluna e imprimindo resultado loja 02
vendas_categoria_produto_LJ02 = vendas_categoria_produto_LJ02.rename(columns={"Preço": "Quantidade_vendida"})
print (f"Resultado da loja 02:\n \n {vendas_categoria_produto_LJ02}")
```

```python
#Loja 03:
# Agrupando por categoria e contando quantidade de vendas
vendas_por_categoria_produto = loja3.groupby("Categoria do Produto")["Preço"].count().reset_index()

# Renomeando a coluna
vendas_por_categoria_produto = vendas_por_categoria_produto.rename(columns={"Preço": "Quantidade_vendida"})

# Exibindo o resultado
print (f"Resultado da loja 03:\n \n {vendas_por_categoria_produto}")
```

```python
#Loja 04:
# Agrupando por categoria e contando quantidade de vendas
vendas_por_categoria_produto = loja4.groupby("Categoria do Produto")["Preço"].count().reset_index()

# Renomeando a coluna
vendas_por_categoria_produto = vendas_por_categoria_produto.rename(columns={"Preço": "Quantidade_vendida"})

# Exibindo o resultado
print (f"Resultado da loja 04:\n \n {vendas_por_categoria_produto}")
```

```python
vendas_por_categoria_produto = vendas_por_categoria_produto.sort_values(by="Quantidade_vendida", ascending=False)
vendas_por_categoria_produto
```
